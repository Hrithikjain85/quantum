Implementing boolean function:
## Q = (A nand B) xnor (A xor B)

```python
import qiskit
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
from qiskit import *
from qiskit.visualization import plot_histogram
import numpy as np
%matplotlib inline
A=QuantumRegister(1,'a')
B=QuantumRegister(1,'b')
W=QuantumRegister(1)
out=ClassicalRegister(1,'result')
qc = QuantumCircuit(A,B,W,out)
qc.x(2)
qc.ccx(A[0],B[0],2)
qc.cx(A[0],B[0])
qc.cx(B[0],2)
qc.x(2)
qc.measure(2,out[0])
qc.draw(output='mpl')
```

```python
def bool1(x,y):
    A=QuantumRegister(1,'a')
    B=QuantumRegister(1,'b')
    W=QuantumRegister(1)
    out=ClassicalRegister(1,'result')
    qc = QuantumCircuit(A,B,W,out)
    if x==1:
        qc.x(0)
    if y==1:
        qc.x(1)
    qc.x(2)
    qc.ccx(A[0],B[0],2)
    qc.cx(A[0],B[0])
    qc.cx(B[0],2)
    qc.x(2)
    qc.measure(2,out[0])
    simulator = Aer.get_backend('qasm_simulator')
    result = execute(qc, backend=simulator).result()
    return plot_histogram(result.get_counts(qc))
```

```python
def BOOL1(x,y):
    A=QuantumRegister(1,'a')
    B=QuantumRegister(1,'b')
    W=QuantumRegister(1)
    out=ClassicalRegister(1,'result')
    qc = QuantumCircuit(A,B,W,out)
    if x==1:
        qc.x(0)
    if y==1:
        qc.x(1)
    qc.x(2)
    qc.ccx(A[0],B[0],2)
    qc.cx(A[0],B[0])
    qc.cx(B[0],2)
    qc.x(2)
    qc.measure(2,out[0])
    qcomp = provider.get_backend('ibmq_bogota')
    job = execute(qc,backend=qcomp)
    from qiskit.tools.monitor import job_monitor
    job_monitor(job)
    result= job.result()
    return plot_histogram(result.get_counts(qc)) 
```
