Implementing boolean function:
## Q = (A and B) or (A xor B)

```python
import qiskit
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
from qiskit import *
from qiskit.visualization import plot_histogram
import numpy as np
%matplotlib inline
A=QuantumRegister(1,'a')
B=QuantumRegister(1,'b')
W=QuantumRegister(3)
out=ClassicalRegister(1,'result')
qc = QuantumCircuit(A,B,W,out)
qc.ccx(A[0],B[0],2)
qc.x(A[0])
qc.x(B[0])
qc.ccx(A[0],B[0],3)
qc.x(2)
qc.x(3)
qc.x(4)
qc.ccx(2,3,4)
qc.measure(4,out[0])
qc.draw(output='mpl')
```

```python
def bool2(x,y):
    A=QuantumRegister(1,'a')
    B=QuantumRegister(1,'b')
    W=QuantumRegister(3)
    out=ClassicalRegister(1,'result')
    qc = QuantumCircuit(A,B,W,out)
    if x==1:
        qc.x(0)
    if y==1:
        qc.x(1)
    qc.ccx(A[0],B[0],2)
    qc.x(A[0])
    qc.x(B[0])
    qc.ccx(A[0],B[0],3)
    qc.x(2)
    qc.x(3)
    qc.x(4)
    qc.ccx(2,3,4)
    qc.measure(4,out[0])
    simulator = Aer.get_backend('qasm_simulator')
    result = execute(qc, backend=simulator).result()
    return plot_histogram(result.get_counts(qc))
```

```python
def BOOL2(x,y):
    A=QuantumRegister(1,'a')
    B=QuantumRegister(1,'b')
    W=QuantumRegister(3)
    out=ClassicalRegister(1,'result')
    qc = QuantumCircuit(A,B,W,out)
    if x==1:
        qc.x(0)
    if y==1:
        qc.x(1)
    qc.ccx(A[0],B[0],2)
    qc.x(A[0])
    qc.x(B[0])
    qc.ccx(A[0],B[0],3)
    qc.x(2)
    qc.x(3)
    qc.x(4)
    qc.ccx(2,3,4)
    qc.measure(4,out[0])
    qcomp = provider.get_backend('ibmq_bogota')
    job = execute(qc,backend=qcomp)
    from qiskit.tools.monitor import job_monitor
    job_monitor(job)
    result= job.result()
    return plot_histogram(result.get_counts(qc)) 
```
