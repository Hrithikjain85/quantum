```python
import qiskit
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
from qiskit import *
from qiskit.visualization import plot_histogram
import numpy as np
%matplotlib inline
A=QuantumRegister(1,'a')
B=QuantumRegister(1,'b')
C=QuantumRegister(1,'c')
W=QuantumRegister(1)
V=QuantumRegister(1)
out=ClassicalRegister(1,'result')
qc = QuantumCircuit(A,B,C,W,V,out)
qc.x(3)
qc.ccx(A[0],B[0],3)
qc.x(3)
qc.x(4)
qc.ccx(3,C[0],4)
qc.measure(V[0], out[0])
qc.draw(output='mpl')
```

```python
def nand3(x,y,z):
    A=QuantumRegister(1,'a')
    B=QuantumRegister(1,'b')
    C=QuantumRegister(1,'c')
    W=QuantumRegister(1)
    V=QuantumRegister(1)
    out=ClassicalRegister(1,'result')
    qc = QuantumCircuit(A,B,C,W,V,out)
    if x==1:
        qc.x(A[0])
    if y==1:
        qc.x(B[0])
    if z==1:
        qc.x(C[0])
    qc.x(3)
    qc.ccx(A[0],B[0],3)
    qc.x(3)
    qc.x(4)
    qc.ccx(3,C[0],4)
    qc.measure(V[0], out[0])
    simulator = Aer.get_backend('qasm_simulator')
    result = execute(qc, backend=simulator).result()
    return plot_histogram(result.get_counts(qc)) 
```

```python
def NAND3(x,y,z):
    A=QuantumRegister(1,'a')
    B=QuantumRegister(1,'b')
    C=QuantumRegister(1,'c')
    W=QuantumRegister(1)
    V=QuantumRegister(1)
    out=ClassicalRegister(1,'result')
    qc = QuantumCircuit(A,B,C,W,V,out)
    if x==1:
        qc.x(A[0])
    if y==1:
        qc.x(B[0])
    if z==1:
        qc.x(C[0])
    qc.x(3)
    qc.ccx(A[0],B[0],3)
    qc.x(3)
    qc.x(4)
    qc.ccx(3,C[0],4)
    qc.measure(V[0], out[0])
    qcomp = provider.get_backend('ibmq_manila')
    job = execute(qc,backend=qcomp)
    from qiskit.tools.monitor import job_monitor
    job_monitor(job)
    result= job.result()
    return plot_histogram(result.get_counts(qc)) 
```
